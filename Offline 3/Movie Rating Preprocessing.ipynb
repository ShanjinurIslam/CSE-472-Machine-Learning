```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('ratings.csv')
df = df.drop(columns=df.columns[-1])
```

```python
X = np.zeros((max(df['userId']),max(df['movieId'])))
```

```python
X
```

```python
input_matrix = df.values
```

```python
for each in input_matrix:
    X[int(each[0])-1][int(each[1])-1] = each[2]
```

```python
df = pd.DataFrame(data=X)
```

```python
df
```

```python
df.to_csv('movie_dataset.csv',index=False)
```
